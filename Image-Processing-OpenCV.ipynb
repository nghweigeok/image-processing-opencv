# Load Image and Convert Color Channels

```python
img_greyscale = cv2.imread('./photo.jpg', 0)
img_greyscale.shape
```

```python
img = cv2.imread('./photo.jpg')
```

```python
img.shape  # (height, width, channels)
```

```python
# OpenCV default BGR channels
plt.imshow(img)
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

# Resize

```python
img_RGB.shape
```

```python
width = int(img_RGB.shape[1] / 2)
height = int(img_RGB.shape[0] / 2)

img_RGB_smaller = cv2.resize(src=img_RGB, dsize=(width, height)) 
img_RGB_smaller.shape
```

```python
plt.imshow(img_RGB_smaller)
```

```python
width = int(img_RGB.shape[1] * 2)
height = int(img_RGB.shape[0] * 2)

img_RGB_bigger = cv2.resize(src=img_RGB, dsize=(width, height)) 
img_RGB_bigger.shape
```

```python
plt.imshow(img_RGB_bigger)
```

# Blending

```python
img_overlay = cv2.imread('./photo-overlay.jpg')
img_overlay.shape
```

```python
plt.imshow(img_overlay)
```

### Photo by <a href="https://unsplash.com/@morvanic?utm_source=unsplash&utm_medium=referral&utm_content=creditCopyText">Morvanic Lee</a> on <a href="/s/photos/quote?utm_source=unsplash&utm_medium=referral&utm_content=creditCopyText">Unsplash</a>
  

```python
# make them the same size

img_overlay = cv2.resize(img_overlay, (img_RGB.shape[1], img_RGB.shape[0]))
img_overlay.shape
```

```python
plt.imshow(img_overlay)
```

```python
blended = cv2.addWeighted(src1=img_RGB, alpha=0.3, src2=img_overlay, beta=0.7, gamma=0)
```

```python
plt.imshow(blended)
```

# Blurring

```python
img_small = img_RGB[4000:, 2000:]
plt.imshow(img_small)
print(img_small.shape)
```

```python
blurred = cv2.blur(src=img_small, ksize=(100, 100))
```

```python
plt.imshow(blurred)
```

# Thresholding

```python
plt.imshow(img_greyscale, cmap='gray')
```

```python
# greyscale
ret, thresh1 = cv2.threshold(src=img_greyscale, thresh=127, maxval=255, type=cv2.THRESH_BINARY)
```
